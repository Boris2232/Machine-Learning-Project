```python
import pandas as pd
from matplotlib import pyplot as plt



fig,axes=plt.subplots(1,2,figsize=(10,7))
fig.tight_layout(pad=6.0)
fig.suptitle("UTK", fontsize=15)
colors = ['cadetblue', 'violet', 'blue', 'green', 'yellow','cyan','magenta','maroon','coral']

with open('UTK_faces.csv', 'r') as file:
    Utk_faces = pd.read_csv(file, delimiter=';')
x = Utk_faces['gender']
x = x.replace([0.0, 1.0], ['Male', 'Female'])
y = x.value_counts()
explode = [0, 0.1]
axes[0].pie(y, labels=['Male', 'Female'], shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[0].axis('equal')
axes[0].legend(loc='upper left', title='Genders', fontsize='large')
#



x = Utk_faces['race']
x = x.replace([i for i in range(5)], ['White', 'Black', ' Asian', 'Indian', 'Others']).value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[1].pie(counter, labels=labels, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[1].axis('equal')
axes[1].legend(loc='upper left', title='Race', fontsize='large')
plt.show()
#
#
#
#
# Fair-face train split
fig, axes = plt.subplots(1, 3, figsize=(15, 7))
fig.tight_layout(pad=6.0)
fig.suptitle("Fair-Face train split", fontsize=15)

with open('Fair-Face.csv', 'r') as file:
    Fair_Face = pd.read_csv(file, delimiter=';')
# doing with train split
x = Fair_Face['gender'].loc[Fair_Face['split'] == 'train']
y = x.value_counts()
explode = [0, 0.1]
axes[0].pie(y, labels=['Male', 'Female'], shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[0].axis('equal')
axes[0].legend(loc='upper left', title='Genders', fontsize='large')



x = Fair_Face['race'].loc[Fair_Face['split'] == 'train'].value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[1].pie(counter, labels=labels, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[1].axis('equal')
axes[1].legend(loc='lower center', title='Race', fontsize='small')


x = Fair_Face['age_range'].loc[Fair_Face['split'] == 'train'].value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[2].pie(counter, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=65)
axes[2].axis('equal')
axes[2].legend(loc='upper left', title='Age groups', fontsize='small',labels=labels)
plt.show()
#
#
#
#

# Fair-face val split
fig, axes = plt.subplots(1, 3, figsize=(15, 7))
fig.tight_layout(pad=6.0)
fig.suptitle("Fair-Face val split", fontsize=15)

# doing with val split
x = Fair_Face['gender'].loc[Fair_Face['split'] == 'val']
y = x.value_counts()
explode = [0, 0.1]
axes[0].pie(y, labels=['Male', 'Female'], shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[0].axis('equal')
axes[0].legend(loc='upper left', title='Genders', fontsize='large')



x = Fair_Face['race'].loc[Fair_Face['split'] == 'val'].value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[1].pie(counter, labels=labels, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[1].axis('equal')
axes[1].legend(loc='lower center', title='Race', fontsize='small')


x = Fair_Face['age_range'].loc[Fair_Face['split'] == 'val'].value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[2].pie(counter, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=65)
axes[2].axis('equal')
axes[2].legend(loc='upper left', title='Age groups', fontsize='small',labels=labels)
plt.show()
#
#
#
#
#
# RAF_DB train split
fig, axes = plt.subplots(1, 4, figsize=(15, 7))
fig.tight_layout(pad=6.0)
fig.suptitle("RAF_DB train split", fontsize=15)
with open('RAF_DB.csv', 'r') as file:
    RAF_DB = pd.read_csv(file, delimiter=';')

# doing with train split
x = RAF_DB['gender'].loc[RAF_DB['split'] == 'train'].replace([1, 2,3], ['Male', 'Female','Unsure'])
y = x.value_counts()
explode = [0.05 for i in range(len(y))]
axes[0].pie(y, labels=['Male', 'Female','Unsure'], shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[0].axis('equal')
axes[0].legend(loc='upper left', title='Genders', fontsize='large')



x = RAF_DB['race'].loc[RAF_DB['split'] == 'train'].replace([0, 1, 2],
                                                           ['Caucasian', 'African-American', 'Asian']).value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[1].pie(counter, labels=labels, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[1].axis('equal')
axes[1].legend(loc='lower center', title='Race', fontsize='small')


x = RAF_DB['age_range'].loc[RAF_DB['split'] == 'train'].replace([0, 1, 2,3,4],
                                                           ['0-3', '4-19', '20-39', '40-69', '70+']).value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[2].pie(counter, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=65)
axes[2].axis('equal')
axes[2].legend(loc='upper left', title='Age groups', fontsize='small',labels=labels)


x = RAF_DB['emotion'].loc[RAF_DB['split'] == 'train'].replace([1, 2,3,4,5,6,7],
                                                           ['Surprise', 'Fear', 'Disgust', 'Happiness', 'Sadness', 'Anger', 'Neutral']).value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[3].pie(counter, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=65)
axes[3].axis('equal')
axes[3].legend(loc='upper left', title='Emotions', fontsize='small',labels=labels)


# RAF_DB test split
fig, axes = plt.subplots(1, 4, figsize=(15, 7))
fig.tight_layout(pad=6.0)
fig.suptitle("RAF_DB test split", fontsize=15)

# doing with test split
x = RAF_DB['gender'].loc[RAF_DB['split'] == 'test'].replace([1, 2,3], ['Male', 'Female','Unsure'])
y = x.value_counts()
explode = [0.05 for i in range(len(y))]
axes[0].pie(y, labels=['Male', 'Female','Unsure'], shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[0].axis('equal')
axes[0].legend(loc='upper left', title='Genders', fontsize='large')



x = RAF_DB['race'].loc[RAF_DB['split'] == 'test'].replace([0, 1, 2],
                                                           ['Caucasian', 'African-American', 'Asian']).value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[1].pie(counter, labels=labels, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=90)
axes[1].axis('equal')
axes[1].legend(loc='lower center', title='Race', fontsize='small')


x = RAF_DB['age_range'].loc[RAF_DB['split'] == 'test'].replace([0, 1, 2,3,4],
                                                           ['0-3', '4-19', '20-39', '40-69', '70+']).value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[2].pie(counter, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=65)
axes[2].axis('equal')
axes[2].legend(loc='upper left', title='Age groups', fontsize='small',labels=labels)


x = RAF_DB['emotion'].loc[RAF_DB['split'] == 'test'].replace([1, 2,3,4,5,6,7],
                                                           ['Surprise', 'Fear', 'Disgust', 'Happiness', 'Sadness', 'Anger', 'Neutral']).value_counts()
labels = list(x.index)
counter = list(x.values)
explode = [0.05 for i in range(len(labels))]
axes[3].pie(counter, shadow=True, autopct='%1.1f%%', colors=colors, explode=explode, startangle=65)
axes[3].axis('equal')
axes[3].legend(loc='upper left', title='Emotions', fontsize='small',labels=labels)
plt.show()
```
