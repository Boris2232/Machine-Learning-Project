```python
import os

from PIL import Image

in_folder = './YOUR_INPUT_FOLDER'
out_folder = './YOUR_OUTPUT_FOLDER'
# to have readable images name
counter=0
for i in os.listdir(in_folder):
    counter+=1
    # Load the original image
    midjourney_image = Image.open(f'{in_folder}/{i}')
    # Split the image into four equal parts
    width, height = midjourney_image.size
    split_width = width // 2
    split_height = height // 2

    # Split and save the images
    top_left = midjourney_image.crop((0, 0, split_width, split_height))
    top_left.save(f"./{out_folder}/cropped_top_left_{counter}.jpg")

    top_right = midjourney_image.crop((split_width, 0, width, split_height))
    top_right.save(f"./{out_folder}/cropped_top_right_{counter}.jpg")

    bottom_left = midjourney_image.crop((0, split_height, split_width, height))
    bottom_left.save(f"./{out_folder}/cropped_bottom_left_{counter}.jpg")

    bottom_right = midjourney_image.crop((split_width, split_height, width, height))
    bottom_right.save(f"./{out_folder}/cropped_bottom_right_{counter}.jpg")
```
